```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
file_path = '/content/ingresos-1.txt'  # Cambiar por la ruta local o en Colab
df = pd.read_csv(file_path, sep='\t')  # Asegurarse de que el separador es el correcto

# Inspección inicial
print("Primeras filas del dataset:")
print(df.head())

print("\nResumen estadístico:")
print(df.describe())

print("\nInformación del dataset:")
print(df.info())
```

```python
# Verificar valores faltantes
missing_values = df.isnull().sum()
print("\nValores faltantes por columna:")
print(missing_values)
```

```python
# Imputar valores faltantes para variables numéricas con la mediana
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Para variables categóricas, se usa la moda
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nRevisión después de imputación:")
print(df.isnull().sum())
```

```python
# Estadísticas descriptivas para las variables numéricas
print("\nEstadísticas descriptivas para variables numéricas:")
print(df[numerical_cols].describe())

# Estadísticas para variables categóricas
print("\nDistribución de valores para variables categóricas:")
for col in categorical_cols:
    print(f"{col}:\n{df[col].value_counts()}\n")
```

```python
df[numerical_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle("Histogramas de Variables Numéricas", fontsize=16)
plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f"Distribución de {col}")
    plt.show()
```

```python
# Revisar valores únicos en 'etnia' antes de la transformación
print("Valores únicos en 'etnia':")
print(df['etnia'].unique())

# Transformar 'etnia' a variables dummy (one-hot encoding)
df = pd.get_dummies(df, columns=['etnia'], drop_first=True)

# Verificar columnas transformadas
print("Columnas después de aplicar One-Hot Encoding a 'etnia':")
print(df.columns)

# Transformar 'genero' a valores binarios
df['genero'] = df['genero'].map({'masculino': 1, 'femenino': 0})

# Verificar valores únicos en 'genero'
print("\nValores únicos en 'genero' después de la transformación:")
print(df['genero'].unique())
```

```python
# Matriz de correlación
correlation_matrix = df.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Matriz de Correlación")
plt.show()
```

```python
# Boxplot para ingresos vs. género
plt.figure(figsize=(8, 6))
sns.boxplot(x='genero', y='ingresos', data=df)
plt.title("Ingresos por Género")
plt.show()

# Scatterplot para ingresos vs. educación
plt.figure(figsize=(8, 6))
sns.scatterplot(x='educacion', y='ingresos', data=df)
plt.title("Ingresos vs. Años de Educación")
plt.show()
```
